### Code to compare two groups using Kruschke's BEST method

_author_: Ata Kalirad

John Kruschke proposed an alternative to t-test, dubbed BEST (Bayesian estimation supersedes the t test). BEST proveds bayesian estimates for **2** groups, and provides credible values for the effect size, group means and their difference, standard deviations and their difference, and the normality of the data. For detailed information on this method, see Kruschke, J. K. (2013). [Bayesian estimation supersedes the t test. Journal of Experimental Psychology: General, 142(2), 573â€“603](https://psycnet.apa.org/doi/10.1037/a0029146).

- The structure of the model:

<img src="../supp_data/BEST_model.png" alt="Drawing" style="width: 500px;"/>

In this model, datum $i$ from group $j$ ($y_{i,j}$) is drawn from Student's t-distribution. The tail of the distribution is controlled by parameter $\nu$. As $\nu \rightarrow \infty$, the t-distribution becomes the normal distribution. $\mu$ and $\sigma$ determine the location and the scale of the distribution.

```python
x = np.linspace(-5, 5, 100)

for i in [1, 2, 10]:
    y = stats.t(df=i, loc=0, scale=1).pdf(x)
    plt.plot(x, y, label=r'$\nu=$' + str(i))
plt.ylabel(r'$f(x)$')
plt.xlabel(r'$x$')
plt.title(r'The effect of $\nu$ on the t-distribution ($\mu=0$, $\sigma=1$)')
plt.legend();
```

To illustrate the model, we generate two random samples drawn: 

_Note_: Samples do not need to be of equal size.

```python
sample_1 = np.random.poisson(10, size=50)
sample_2 = np.random.poisson(5, size=50)
```

```python
fig, ax = plt.subplots(
    nrows=1,
    ncols=1,
    figsize=(5, 5))

plot = ax.boxplot([sample_1, sample_2])

ax.set(xticklabels=['Group 1', 'Group 2']);
```

```python
def best_two_groups(sample_1, sample_2, t_tune=1000):
    pooled = np.concatenate([sample_1, sample_2])
    m_s = np.mean(pooled)
    std_s = np.std(pooled) * 2
    Ïƒ_low = 1
    Ïƒ_high = 300
    n1 = len(sample_1)
    n2 = len(sample_2)
    with pm.Model() as model:
        # Î¼1 and Î¼2 for each group is drawn from Normal dist.
        group1_mean = pm.Normal("group1_mean", mu=m_s, sigma=std_s)
        group2_mean = pm.Normal("group2_mean", mu=m_s, sigma=std_s)
        # Ïƒ1 and Ïƒ2 for each group is drawn from uniform dist.
        group1_std = pm.Uniform("group1_std", lower=Ïƒ_low, upper=Ïƒ_high)
        group2_std = pm.Uniform("group2_std", lower=Ïƒ_low, upper=Ïƒ_high)
        # degrees of freedom of the Student's t-distribution is drawn from exponential distribution
        Î½ = pm.Exponential("Î½_minus_one", 1 / 29.0) + 1
        Î»1 = group1_std ** -2
        Î»2 = group2_std ** -2
        group1 = pm.StudentT("observed1", nu=Î½, mu=group1_mean, lam=Î»2, observed=sample_1)
        group2 = pm.StudentT("observed2", nu=Î½, mu=group2_mean, lam=Î»1, observed=sample_2)
        diff_of_means = pm.Deterministic("difference of means", group1_mean - group2_mean)
        diff_of_stds = pm.Deterministic("difference of stds", group1_std - group2_std)
        effect_size = pm.Deterministic(
            "effect size", diff_of_means / np.sqrt((group1_std ** 2 * (n1 - 1) + group2_std ** 2 * (n2 - 1)) / (n1 + n2 - 2)))
        trace = pm.sample(5000, tune=t_tune, random_seed=RANDOM_SEED)
    return trace
```

```python
test = best_two_groups(sample_1, sample_2)
```

```python
az.plot_trace(test, var_names=['group1_mean', 'group2_mean'], compact=False);
```

The results of the model are quite encouraging for a couple of reasons:

- All the four Markov chains have converged during sampling, as evident by the similarity of the estimated distributions for all the parameters of the model. This point can be quantitatively shown by using the Gelman-Rubin statistics ($\hat{R}$), which indicates convergence when $\hat{R}$ is close to 1.

```python
az.summary(test, var_names=['group1_mean', 'group2_mean'], hdi_prob=0.95, round_to=2)
```

- The posterior distribution for the difference in mean correctly inferred the difference between our two groups, which lies beyond an arbitrary region of practical equivalence within $[-2, +2]$. 

```python
az.plot_posterior(test, var_names=['difference of means'], hdi_prob=0.95, rope=[-2, 2]);
```

Our model calculates the effect size as well, using Hedges' formulation (1981):

$\delta = \frac{\mu_1 - \mu_2}{ \sqrt{\frac{\sigma^2_1(n_1 - 1) + \sigma^2_2(n_2 - 1)}{n_1 + n_2 - 2}}}$

```python
az.plot_posterior(test, var_names=['effect size'], hdi_prob=0.95);
```

The bulk of the posterior distribution for the effect size is greater than 1.2.

In addition to estimating the difference in means, BEST estimates the difference in stardust deviation between the samples as well. This ability exceeds the what the usual t-test can achieve:

```python
az.plot_posterior(test, var_names=['difference of stds'], hdi_prob=0.95);
```

The inferred difference in standard deviations between the two groups is logical since the samples were drawn from $\mathrm{Poisson}(\lambda)$ , where $\sigma = \lambda$.
